```python
import numpy as np
import pandas as pd
```

```python
dataf = pd.DataFrame( np.random.randint(1,100, size = (5,4)) ,columns = ['A','B','C','D'])
dataf_1 = pd.DataFrame( np.random.randint(1,100, size = (5,4)) ,columns = ['A','B','C','D'])
dataf_2 = pd.DataFrame( np.random.randint(1,100, size = (5,4)) ,columns = ['E','F','G','H'])
```

```python
dataf
```

```python
dataf_1
```

```python
dataf_2
```

```python
#Funkcja concat, łączy dataframes jeden pod drugim.
df_con = pd.concat([dataf,dataf_1])
df_con
```

```python
#Jak widzimy na górze mamy kilka tych samych wartości w indexach.
#Musimy teraz zresetować indeksy. 
df_con.reset_index(drop = True)
```

```python
#Funkcja join scala dataframes obok siebie. 
dataf.join(dataf_2)
```

```python
#Merge działa podobnie do join. Przydaje się jednak, jak chcemy połączyć DataFrames
#na podstawie wartości w kolumnach, a nie indexu.
df_merged = pd.merge(dataf, dataf_2,left_index = True, right_index = True)
df_merged
```

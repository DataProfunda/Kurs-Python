```python
import numpy as np
import pandas as pd
```

```python
# Pandas to biblioteka oparta na Numpy. Jego głównym obiektem jest DataFrame,
#tak jak w bibliotece Numpy jest tablica. 
#DataFrame różni się tym, że są nazwy kolumn,
#oraz DataFrame może mieć różne typy(String Int czy Bool) jednocześnie.
#W konkretnej kolumnie muszą być wartości tego samego typu.
```

```python
#Z list stworzyliśmy dataframe
pracownik1 = ['Jan','Kowalski',34]
pracownik2 = ['Adam', 'Nowak',52]
pracownik3 = ['Marcin','Adamczyk',18]
pracownik4 = ['Wojtek','Suchodolski',29] 
pracownik5 = ['Marek','Lewandowski',18]
kolumny = ['Imie','Nazwisko','Wiek']
df = pd.DataFrame( [pracownik1 ,pracownik2 ,pracownik3 ,pracownik4,pracownik5], columns = kolumny) 
df
```

```python
#Wpisanie nazwy kolumny w [] wyświetla nam całą kolumnę.
df['Imie']
```

```python
#Jak widzimy Seria, to po prostu jedna kolumna.
type(df['Imie'])
```

```python
#Możemy wyświetlać także kilka kolumn.
#Zauważmy że w [] podajemy listę kolumn, do wyświetlenia.
df[ ['Imie','Nazwisko'] ]
```

```python
#Możemy wyświetlać konkretne wiersze
#dataframe.loc[ nr_wiersza , 'nazwa kolumny' ]
#W tym przypadku wyświetliliśmy wiersz pierwszy.
# ':'' oznacza, że chcemy wyświetlić wszystkie kolumny.
df.loc[1,:]
```

```python
df.loc[1, ['Imie','Wiek'] ]
```

```python
#iloc służy do wyświetlania wartości za pomocą indeksów kolumn, a nie poprzez nazwy.
df.iloc[1,[0,1]]
```

```python
#W tym przypadku wyświetliliśmy kolumnę nr 1 i 2 od wiersza nr 0 do 4.
df.iloc[0:4,[1,2]]
```

```python
#W tym przypadku wyświetliliśmy kolumny od 0 do 2 w wierszu nr 2.
df.iloc[2,0:2]
```

```python
#Jeżeli chcemy stworzyć nową kolumnę, 
#odwołujemy się jak do istniejącej kolumny, jednak wpisujemy nową nazwę.
df['Pensja'] = np.random.randint(3500,10000,5)
df
```

```python
#Jeżeli chcemy dodać nowy wiersz, 
#odwołujemy się do nowego indexu.
df.loc[5,:] = ['Tomek','Kaminski',22,6700]
df
```

```python
#Funkcja drop() służy do usunięcia wiersza lub kolumny.
#Domyślnie axis=0, czyli usuwamy wiersz nr 2.
df.drop(2)
```

```python
#Jeżeli w funkcji drop() axis=1 mówimy, że chcemy usunąć kolumnę.  
df.drop('Nazwisko', axis=1)
```

```python
#Jeżeli w miejsce indexu wpiszemy warunek,
#wyświetlą nam się tylko te wiersze, które spełniają ten warunek.
df.loc[ df['Pensja'] > 7000 , :]
```

```python
#Jeżeli chcemy, żeby były spełnione dwa warunki, łączymy te warunki znakiem ' & ',
#Jest on odpowiednikiem łącznika 'and'.
df.loc[ (df['Pensja'] > 5000) & (df['Wiek'] > 20 ), :]
```

```python
#Nie możemy w przypadku wielu wierszy stosować łącznika 'and', ponieważ służy on tylko
#do pojedyńczych wartości True lub False.
df.loc[ (df['Pensja'] > 5000) and (df['Wiek'] > 20 ), :]
```

```python
#Znak ' | ', jest odpowiednikiem słowa 'or'.
df.loc[ (df['Pensja'] > 8000) | (df['Wiek'] > 35 ), :]
```

```python
df = pd.DataFrame( np.random.randint(1,100, size = (8,6)) ,columns = ['A','B','C','D','E','F'])
df
```

```python
# np.nan , oznacza wartość nie będąca liczbą. NaN pochodzi z  'Not A Number'.
#Najczęściej oznacza się tak niewiadome wartości.
df.loc[2,['C','D']] = np.nan
df.loc[1,'D'] = np.nan
df
```

```python
#Funkcja dropna(), usuwa te wiersze, gdzie są wartości Nan.
#W skrócie Drop 'Not a Number'.
df.dropna()
```

```python
#Jeżeli chcemy usunąć kolumny, tak gdzie są wartości NaN, oznaczamy axis=1.
#Domyślnie axis=0.
df.dropna(axis=1)
```

```python
#Funkcja fillna(arg), wypełnia wszystkie wartości NaN, wartością podaną w funkcji.
#Jak widzimy tam gdzie było NaN jest teraz 13.
df.fillna(13)
```

```python
#Funkcja mean() zwraca nam średnie wartości kolumn.
df.mean()
```

```python
#Jeżeli do funkcji fillna(), jako argument wpiszemy wartości z funkcji mean(), 
#zamienia NaN na wartość średnią w danej kolumnie.
df.fillna( df.mean() )
```

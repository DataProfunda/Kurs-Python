```python
import numpy as np
import pandas as pd
```

```python
# Z list stworzyliśmy dataframe
pracownik1 = ['Jan','Kowalski',34]
pracownik2 = ['Adam', 'Nowak',52]
pracownik3 = ['Marcin','Adamczyk',18]
pracownik4 = ['Wojtek','Suchodolski',29] 
pracownik5 = ['Marek','Lewandowski',18]
kolumny = ['Imie','Nazwisko','Wiek']
df = pd.DataFrame( [pracownik1 ,pracownik2 ,pracownik3 ,pracownik4,pracownik5], columns = kolumny) 
df['Pensja'] = np.random.randint(3500,10000,5)
df
```

```python
#Funkcja sum() sumuje wszystkie wartości z kolumny 'Pensja'
df['Pensja'].sum()
```

```python
#Funkcja mean() wyświetla średnią z kolumny 'Pensja'.
df['Pensja'].mean()
```

```python
#Dla dowodu, że funkcja mean() działa, podzielmy sumę pensji przez ilość osób.
srednia = df['Pensja'].sum() / len(df['Pensja'])
srednia
```

```python
#Funkcja unique() wyświetla unikalne wartości w kolumnie.
#Wyświetliły się 4 wartości, ponieważ 18 jest w dwóch miejscach.
df['Wiek'].unique()
```

```python
df = pd.DataFrame( np.random.randint(1,100, size = (8,6)) ,columns = ['A','B','C','D','E','F'])
df
```

```python
def szescian(a):
    return a * a * a
```

```python
#Jeżeli chcemy zastosować jakąś funcję użwamy funkcji apply(funkcja)
df.apply(szescian)
```

```python
#Na podstawie jednej z kolumn możemy posortować wiersze
df.sort_values(by='C')
```

```python
#Jeżeli chcemy posortować wartości malejąco, dodajemy ascending=False.
#Domyślnie ascending=True.
df.sort_values(by='C',ascending=False)
```
